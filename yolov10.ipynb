```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'F:/TEJA/IP/PROJ_3/runs/detect/train3/weights/best.pt')

# Video input and output paths
input_video_path = r"F:\TEJA\DL\PROJ_3\dataset\video2.mov"
output_video_path = r"F:\TEJA\DL\PROJ_3\dataset\output_video12.mov"

# Load video
cap = cv2.VideoCapture(input_video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform detection
    results = model.predict(frame)

    # Annotate detections
    for result in results[0].boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = result
        x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
        label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

        # Draw bounding box and label
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write frame to the output video
    out.write(frame)

    # Optional: Display the frame (for debugging)
    cv2.imshow('Pomegranate Stage Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'F:/TEJA/IP/PROJ_3/runs/detect/train3/weights/best.pt')

# Open system's camera
cap = cv2.VideoCapture(0)  # 0 for default camera

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open the camera.")
    exit()

# Get video properties (optional)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Process video frame by frame
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read the frame.")
        break

    # Perform detection
    results = model.predict(frame, verbose=False)

    # Annotate detections
    for result in results[0].boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = result
        x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
        label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

        # Draw bounding box and label
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with annotations
    cv2.imshow('Pomegranate Stage Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

## Dataset visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Assuming we have images in each dataset folder and we want to segregate them
train_dir = r'F:\TEJA\DL\PROJ_3\dataset\train\images'
test_dir = r'F:\TEJA\DL\PROJ_3\dataset\test\images'
valid_dir = r'F:\TEJA\DL\PROJ_3\dataset\valid\images'

# Function to count the number of images in each directory
def count_images(directory):
    return len([f for f in os.listdir(directory) if f.endswith('.jpg') or f.endswith('.png')])

# Get the count of images in each dataset
train_count = count_images(train_dir)
test_count = count_images(test_dir)
valid_count = count_images(valid_dir)

# Example 1: Pie chart showing the distribution of images in the datasets
labels = ['Train', 'Test', 'Valid']
sizes = [train_count, test_count, valid_count]
colors = ['red', 'blue', 'green']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Dataset Distribution")
plt.show()

# Example 2: Bar plot showing the count of images in each dataset
datasets = ['Train', 'Test', 'Valid']
counts = [train_count, test_count, valid_count]

plt.figure(figsize=(10, 6))
plt.bar(datasets, counts, color=colors)
plt.title("Image Count in Each Dataset")
plt.xlabel("Dataset")
plt.ylabel("Number of Images")
plt.show()

def get_image_size(directory):
    sizes = []
    for img_file in os.listdir(directory):
        if img_file.endswith('.jpg') or img_file.endswith('.png'):
            img_path = os.path.join(directory, img_file)
            img = plt.imread(img_path)  # Reads the image
            sizes.append(img.shape[0] * img.shape[1])  # width * height
            break  # To consider only the first image
    return sizes

# Assuming you've gathered sizes for train, test, and validation datasets
sizes_train = get_image_size(train_dir)
sizes_test = get_image_size(test_dir)
sizes_valid = get_image_size(valid_dir)

plt.figure(figsize=(10, 6))
plt.hist([sizes_train, sizes_test, sizes_valid], bins=10, color=['red', 'blue', 'green'], label=['Train', 'Test', 'Valid'])
plt.title("Image Size Distribution in Each Dataset")
plt.xlabel("Image Size (pixels)")
plt.ylabel("Frequency")
plt.legend()
plt.show()



# Cumulative counts for all datasets
cumulative_train = np.cumsum([train_count, train_count, train_count])
cumulative_test = np.cumsum([test_count, test_count, test_count])
cumulative_valid = np.cumsum([valid_count, valid_count, valid_count])

plt.figure(figsize=(10, 6))
plt.plot(datasets, cumulative_train, label='Train', color='red', marker='o')
plt.plot(datasets, cumulative_test, label='Test', color='blue', marker='o')
plt.plot(datasets, cumulative_valid, label='Valid', color='green', marker='o')
plt.title("Cumulative Image Count in Each Dataset")
plt.xlabel("Dataset")
plt.ylabel("Cumulative Count of Images")
plt.legend()
plt.show()
```
