```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Paths to the dataset splits
dataset_path = '../dataset/'  # Update this path
splits = ['train', 'valid']
class_names = ['fire','smoke']


def count_classes(split_path, class_names):
    counts = np.zeros(len(class_names), dtype=int)
    label_path = os.path.join(dataset_path, split_path, 'labels')
    for label_file in os.listdir(label_path):
        with open(os.path.join(label_path, label_file), 'r') as file:
            for line in file.readlines():
                class_id = int(line.strip().split()[0])
                if class_id >= len(class_names):  # Check for invalid class IDs
                    print(f"Invalid class ID {class_id} in file {label_file}")
                else:
                    counts[class_id] += 1
    return counts


# Count instances for each split
counts_per_split = {split: count_classes(split, class_names) for split in splits}


```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'../runs/detect/train2/weights/best.pt')

# Input and output image paths
input_image_path = r"F:/TEJA/NLP/PROJ14/images/p12.jpg"
output_image_path = r"F:/TEJA/NLP/PROJ14/images/outputp12.jpg"

# Load the image
image = cv2.imread(input_image_path)

# Perform detection
results = model.predict(image)

# Annotate detections
for result in results[0].boxes.data.tolist():
    x1, y1, x2, y2, conf, cls = result
    x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
    label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

    # Draw bounding box and label
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save the output image
cv2.imwrite(output_image_path, image)

# Optional: Display the image
cv2.imshow('Pomegranate Stage Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'../runs/detect/train/weights/best.pt')

# Open system's camera
cap = cv2.VideoCapture(0)  # 0 for default camera

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open the camera.")
    exit()

# Get video properties (optional)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Process video frame by frame
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read the frame.")
        break

    # Perform detection
    results = model.predict(frame, verbose=False)

    # Annotate detections
    for result in results[0].boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = result
        x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
        label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

        # Draw bounding box and label
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with annotations
    cv2.imshow('Pomegranate Stage Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import numpy as np

# Paths to labels and images
LABELS_DIR = "F:/TEJA/NLP/PROJ14/dataset/train/labels"
IMAGES_DIR = "F:/TEJA/NLP/PROJ14/dataset/train/images"

# Class names (adjust based on dataset.yaml)
CLASS_NAMES = {0: "Fire", 1: "Smoke"}

# Data storage
class_counts = {0: 0, 1: 0}
bbox_sizes = []
aspect_ratios = []
object_counts = []
image_sizes = []
heatmap_data = []

# Debugging info
print(f"🟢 Looking for label files in: {LABELS_DIR}")

# Process each label file
for label_file in os.listdir(LABELS_DIR):
    label_path = os.path.join(LABELS_DIR, label_file)
    
    with open(label_path, "r") as file:
        lines = file.readlines()
    
    print(f"📂 Processing: {label_file}, Objects: {len(lines)}")
    object_counts.append(len(lines))  # Count objects per image

    # Load corresponding image
    img_path = os.path.join(IMAGES_DIR, label_file.replace(".txt", ".jpg"))
    
    if not os.path.exists(img_path): 
        print(f"⚠️ Image missing for: {label_file}")
        continue

    img = cv2.imread(img_path)
    
    if img is None:
        print(f"❌ Error loading image: {img_path}")
        continue

    h, w, _ = img.shape
    image_sizes.append((w, h))

    for line in lines:
        values = line.strip().split()

        if len(values) < 5:
            print(f"⚠️ Skipping malformed line: {values}")
            continue

        class_id, x_center, y_center, width, height = map(float, values[:5])

        # Convert to pixel values
        abs_width = width * w
        abs_height = height * h
        bbox_sizes.append(abs_width * abs_height)
        aspect_ratios.append(abs_width / abs_height if abs_height != 0 else 1)
        class_counts[class_id] += 1
        heatmap_data.append((x_center * w, y_center * h))

print("✅ Data processing complete!")
print(f"📊 Total Images Processed: {len(image_sizes)}")
print(f"🔥 Fire Count: {class_counts[0]}, Smoke Count: {class_counts[1]}")
print(f"📦 Total Objects Detected: {sum(object_counts)}")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ✅ Set font for consistent styling
plt.rcParams["font.family"] = "Arial"

# Paths
LABELS_DIR = "F:/TEJA/NLP/PROJ14/dataset/train/labels"
IMAGES_DIR = "F:/TEJA/NLP/PROJ14/dataset/train/images"
SAVE_DIR = "F:/TEJA/NLP/PROJ14/static/images/charts/"

# Class names
CLASS_NAMES = {0: "Fire", 1: "Smoke"}

# Ensure directory exists
os.makedirs(SAVE_DIR, exist_ok=True)

# Data storage
class_counts = {0: 0, 1: 0}
bbox_sizes = []
aspect_ratios = []
object_counts = []
image_sizes = []
heatmap_data = []

# Process label files
for label_file in os.listdir(LABELS_DIR):
    label_path = os.path.join(LABELS_DIR, label_file)
    
    with open(label_path, "r") as file:
        lines = file.readlines()
    
    object_counts.append(len(lines))

    img_path = os.path.join(IMAGES_DIR, label_file.replace(".txt", ".jpg"))
    
    if not os.path.exists(img_path): 
        print(f"⚠️ Image missing for: {label_file}")
        continue

    img = cv2.imread(img_path)
    
    if img is None:
        print(f"❌ Error loading image: {img_path}")
        continue

    h, w, _ = img.shape
    image_sizes.append((w, h))

    for line in lines:
        values = line.strip().split()

        if len(values) < 5:
            continue

        class_id, x_center, y_center, width, height = map(float, values[:5])

        abs_width = width * w
        abs_height = height * h
        bbox_sizes.append(abs_width * abs_height)
        aspect_ratios.append(abs_width / abs_height if abs_height != 0 else 1)
        class_counts[class_id] += 1
        heatmap_data.append((x_center * w, y_center * h))

# Convert data to numpy arrays
bbox_sizes = np.array(bbox_sizes)
aspect_ratios = np.array(aspect_ratios)
image_sizes = np.array(image_sizes)

# ✅ Ensure data consistency for scatter plot
if len(image_sizes) != len(bbox_sizes):
    min_length = min(len(image_sizes), len(bbox_sizes))
    image_sizes = image_sizes[:min_length]
    bbox_sizes = bbox_sizes[:min_length]

# Function to save plots with the same size
def save_plot(fig, filename):
    fig.set_size_inches(8, 6)  # ✅ Ensuring uniform size
    fig.savefig(SAVE_DIR + filename, bbox_inches="tight")
    plt.close(fig)

# 🔵 Class Distribution (Pie Chart)
fig, ax = plt.subplots()
ax.pie(class_counts.values(), labels=[CLASS_NAMES[k] for k in class_counts], autopct="%1.1f%%", colors=['red', 'gray'])
ax.set_title("🔥 Class Distribution: Fire vs Smoke")
save_plot(fig, "class_distribution.png")

# 🔵 Bounding Box Size Distribution (Histogram)
fig, ax = plt.subplots()
sns.histplot(bbox_sizes, bins=30, kde=True, color="blue", ax=ax)
ax.set_xlabel("Bounding Box Size (pixels)")
ax.set_ylabel("Frequency")
ax.set_title("📏 Bounding Box Size Distribution")
save_plot(fig, "bbox_size_distribution.png")

# 🔵 Object Count Per Image (Bar Chart)
fig, ax = plt.subplots()
sns.histplot(object_counts, bins=range(1, max(object_counts) + 2), kde=False, color="green", ax=ax)
ax.set_xlabel("Objects per Image")
ax.set_ylabel("Number of Images")
ax.set_title("📷 Object Count Per Image")
save_plot(fig, "object_count_per_image.png")

# 🔵 Object Area vs Image Area (Scatter Plot)
fig, ax = plt.subplots()
ax.scatter(image_sizes[:, 0] * image_sizes[:, 1], bbox_sizes, alpha=0.5, color="purple")
ax.set_xlabel("Image Area (pixels)")
ax.set_ylabel("Object Bounding Box Area (pixels)")
ax.set_title("📍 Object Area vs Image Area")
save_plot(fig, "object_area_vs_image_area.png")

# 🔵 Heatmap of Object Locations
if heatmap_data:
    heatmap, xedges, yedges = np.histogram2d(
        [h[0] for h in heatmap_data], [h[1] for h in heatmap_data], bins=50
    )
    fig, ax = plt.subplots()
    cax = ax.imshow(heatmap.T, origin="lower", cmap="hot", interpolation="nearest")
    ax.set_xlabel("X Position")
    ax.set_ylabel("Y Position")
    ax.set_title("🔥 Heatmap of Object Locations")
    fig.colorbar(cax)
    save_plot(fig, "heatmap.png")

# 🔵 Aspect Ratio Distribution
fig, ax = plt.subplots()
sns.histplot(aspect_ratios, bins=30, kde=True, color="orange", ax=ax)
ax.set_xlabel("Aspect Ratio (Width / Height)")
ax.set_ylabel("Frequency")
ax.set_title("📏 Aspect Ratio Distribution")
save_plot(fig, "aspect_ratio_distribution.png")

# 🔵 Class Count Over Dataset Splits
fig, ax = plt.subplots()
sns.barplot(x=[CLASS_NAMES[k] for k in class_counts], y=list(class_counts.values()), palette=['red', 'gray'], ax=ax)
ax.set_xlabel("Class")
ax.set_ylabel("Count")
ax.set_title("📊 Object Count in Train Dataset")
save_plot(fig, "class_count_train.png")

# 🔵 Image Resolution Distribution
fig, ax = plt.subplots()
sns.histplot(image_sizes[:, 0] * image_sizes[:, 1], bins=30, kde=True, color="blue", ax=ax)
ax.set_xlabel("Image Resolution (pixels)")
ax.set_ylabel("Frequency")
ax.set_title("📷 Image Resolution Distribution")
save_plot(fig, "image_resolution_distribution.png")

print(f"✅ All plots are now the same size and saved in: {SAVE_DIR}")
```
